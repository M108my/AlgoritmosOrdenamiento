```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('iteracionesSS.csv')

tamanio = data['Tamanio']
mejor_caso = data['Mejor Caso']
peor_caso = data['Peor Caso']
caso_promedio = data['Caso Promedio']

plt.figure(figsize=(10, 6))

plt.plot(tamanio, mejor_caso, label='Mejor Caso', marker='o')
plt.plot(tamanio, peor_caso, label='Peor Caso', marker='s')
plt.plot(tamanio, caso_promedio, label='Caso Promedio', marker='^')

plt.title('Comparación del número de iteraciones según el tamaño del arreglo')
plt.xlabel('Tamaño del Arreglo')
plt.ylabel('Número de Iteraciones')


plt.legend()

plt.grid(True)

plt.show()
```
